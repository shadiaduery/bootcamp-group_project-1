```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import json
import time
from scipy.stats import linregress
from pprint import pprint

cases_pd = "../Resources/covid_clean_CA_TX.csv"
```

```python
#Reading in data and creating a dataframe 
cases_df = pd.read_csv(cases_pd, encoding="ISO-8859-1")
cases_df.head()
```

```python
#Population per county
total_counties = cases_df['population'].count()
total_counties
```

```python
#Counties by state
county_state = cases_df.set_index('County Name')['State']
county_state
```

```python
#Population per county
county_pop = cases_df.set_index('County Name')['population']
county_pop
```

```python
#Create total column
cases_df['Total'] = cases_df['10/30/2020']
cases_df
```

```python
#Separate counties by size
county_bins = [0, 100000, 500000, 1000000, 100000000000 ]
county_bins_title = ['< 10000', '10001 - 500000', '500001 - 1000000', '> 1000000']
cases_df['county_bins'] = pd.cut(county_pop, county_bins, labels = county_bins_title)

#County groups
county_groups = cases_df.groupby('county_bins')

#Average cases per county bin
total_cases = county_groups['Total']

#Create dataframe
county_groups.loc[:,['Population','Total']]
```
